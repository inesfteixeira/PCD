In order to estimate a regression, import the statsmodels module (https://www.statsmodels.org):

`
import statsmodels.api as sm
model = sm.OLS(y, X).fit()
predictions = model.predict(X) 
model.summary()
`
* y may be a series with data corresponding to the target (or dependent variable)
* X may be a dataframe with data corresponding to the features (or independent variable)

```python
import pandas as pd
```

```python
df=pd.read_excel('https://github.com/masterfloss/data/blob/main/exerciseInt.xlsx?raw=true')
```

```python
df.head()
```

```python
df.shape
```

```python
df=df.dropna()
```

```python
df.shape
```

```python
y=df['Income']
X=df.drop(['country','Income'],axis=1)
```

```python
import statsmodels.api as sm

# Note the difference in argument order
model = sm.OLS(y, X).fit()
# make the predictions by the model
predictions = model.predict(X) 

# Print out the statistics
model.summary()

#durbin watson closer to 2 OK
```
